```python
vowels=0
consonants=0
digits=0 
special=0
s=input("enter a string: ")
for i in s:
    if i.isalpha():
        if i in "aeiou":
            vowels+=1
        else:
            consonants+=1
    elif i.isdigit():
        digits+=1
    else:
        special+=1
print(s)
print("Vowels =", vowels)
print("Consonants =", consonants)
print("Digits =", digits)
print("Special Characters =", special)
```

```python
num = int(input("enter a number: "))
t = num
a = len(str(num))
sum = 0
while t > 0:
    digit = t % 10
    sum += digit ** a
    t //= 10
if sum == num:
    print(num, "is an armstrong number")
else:
    print(num, "is not an armstrong number")
```
