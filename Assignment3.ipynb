```python
#1. Write a program that takes a list of numbers and prints all prime numbers in the list using loops. Do not use any built-in prime-checking functions. 
l=[11,17,22,34,65,79,85,107,103]

for n in l:
    c=0
    for i in range(1,n+1):
        if n%i==0:
            c+=1
    if c==2:
        print(n,end=" ")
```

```python
#3. Write a user-defined function that accepts a list of integers and returns:  
#● Count of even numbers  
#● Sum of odd numbers  
#● Use loops and conditional statements inside the function.

def func(l):
    e=0
    s=0
    for i in l:
        if i%2==0:
            e+=1
        else:
            s+=i
    return e,s

print(func([12,13,14,16,17,21,22]))
```

```python
#5. Create a Student class with attributes: name, marks. Add a method to calculate grade based on marks. 
class Student:
    def __init__(self,n,m):
        self.n=n
        self.m=m

    def grade(self):
        if self.m>=90:
            return "A"
        elif self.m>=75:
            return "B"
        elif self.m>=60:
            return "C"
        else:
            return "D"
        
#6. Create at least two objects and display their grades.          

s1=Student("Ram",95)
s2=Student("Shyam",70)
s3=Student("Mohit",40)

print(s1.grade())
print(s2.grade())
print(s3.grade())
```

```python
#7. Create a BankAccount class with private attributes for balance. Provide methods to deposit and withdraw money. Ensure balance cannot be accessed directly.  

class Bank:
    def __init__(self,b):
        self.__b=b

    def deposit(self,a):
        self.__b+=a

    def withdraw(self,a):
        if a<=self.__b:
            self.__b-=a

    def show(self):
        print(self.__b)

x=Bank(7000)
x.deposit(100)
x.withdraw(450)
x.show()
```

```python
# 10. Write a program that reads numbers from a file and calculates:  
# ● Total  
# ● Average  
# ● Handle the case if the file does not exist.
try:
    f=open("num.txt","r")
    s=0
    c=0
    for i in f:
        s+=int(i)
        c+=1
    print("Total=",s)
    print("Average=",s/c)
    f.close()
except FileNotFoundError:
    print("File not found")
```
